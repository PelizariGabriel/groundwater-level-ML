```python
import pandas as pd
```

```python
earth_tide = pd.read_csv('earth_tide.csv').rename(columns={'datetime': 'date_time'})
wells_func = pd.read_csv('funcionamento_pocos_30min.csv', sep=',').dropna()
wells_func[wells_func.columns.difference(['date_time'])] = wells_func[wells_func.columns.difference(['date_time'])].astype(int)
level_rain_pressure = pd.read_csv('nivel-pocos_chuva_pressao_tratados.csv')

# --- Merge dos DataFrames ---
data = level_rain_pressure.merge(wells_func, on='date_time', how='outer')
data = data.merge(earth_tide, on='date_time', how='outer')

# --- Ordenar pelo datetime ---
data = data.sort_values('date_time').reset_index(drop=True)
```

```python
# SELECIONAR SUBSET PARA MODELO
# --- Lista de colunas que você quer manter ---
cols = [
    'date_time',
    'gmu_level_mH2O_compensado',
    'pressao_mH2O_final',
    'status_IMECC',
    'status_FEF',
    'status_ZOOLOGIA',
    'chuva_mm_30min',
    'earth_tide_nm/s2'
]

    

# --- Converter para datetime (caso ainda não esteja) ---
data['date_time'] = pd.to_datetime(data['date_time'])

# --- Filtrar pelo intervalo de datas ---
mask = (data['date_time'] >= '2022-05-11 15:30:00') & (data['date_time'] <= '2022-07-11 11:00:00')

# --- Aplicar o filtro e selecionar colunas ---
data_subset = data.loc[mask, cols].copy()

# --- Tornar 'date_time' o índice ---
data_subset = data_subset.set_index('date_time')

# --- Visualizar resultado ---
print(data_subset.shape)
data_subset.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, SimpleRNN
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Criar features temporais
# ------------------------------------------------------------
df = data_subset.copy()

# Criar feature booleana: 1 = dia útil (seg–sex), 0 = fim de semana (sáb–dom)
df['week_day'] = (df.index.weekday < 5).astype(int)

# Features adicionais
df['lag_2h'] = df['gmu_level_mH2O_compensado'].shift(4)
df['lag_6h'] = df['gmu_level_mH2O_compensado'].shift(12)
df['lag_12h'] = df['gmu_level_mH2O_compensado'].shift(24)
df['chuva_3h'] = df['chuva_mm_30min'].rolling(6).sum()
df['chuva_24h'] = df['chuva_mm_30min'].rolling(48).sum()
df['chuva_7d'] = df['chuva_mm_30min'].rolling(336).sum()

# Remover linhas com NaN geradas pelos lags e acumulados
df.dropna(inplace=True)

# ------------------------------------------------------------
# 2. Separar target e features
# ------------------------------------------------------------
y = df['gmu_level_mH2O_compensado']
X = df.drop(columns=['gmu_level_mH2O_compensado'])

# Normalizar
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1,1))

# ------------------------------------------------------------
# 3. Criar janelas temporais (para RNN)
# ------------------------------------------------------------
def create_sequences(X, y, window=24, horizon=0):  # horizon=2 → 1h à frente se passo=30min
    Xs, ys = [], []
    for i in range(window, len(X) - horizon):
        Xs.append(X[i-window:i])
        ys.append(y[i + horizon])
    return np.array(Xs), np.array(ys)

window_size = 24  # 24 amostras (~12h se o passo for 30min)
X_seq, y_seq = create_sequences(X_scaled, y_scaled, window_size)

# Divisão treino/teste
train_size = int(len(X_seq) * 0.7)
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

# ------------------------------------------------------------
# 4. Modelo RNN simples
# ------------------------------------------------------------
model = Sequential([
    SimpleRNN(64, activation='relu', return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)

# ------------------------------------------------------------
# 5. Avaliação
# ------------------------------------------------------------
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_true = scaler_y.inverse_transform(y_test)

# Métricas clássicas
r2 = r2_score(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(np.mean((y_true - y_pred)**2))

# NSE (Nash-Sutcliffe Efficiency)
nse = 1 - np.sum((y_true - y_pred)**2) / np.sum((y_true - np.mean(y_true))**2)

print(f"R²: {r2:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"NSE: {nse:.4f}")

# Obter timestamps correspondentes ao conjunto de teste
timestamps = df.index[window_size + train_size : window_size + train_size + len(y_test)]

# ------------------------------------------------------------
# 6. Plotar resultados com datetime
# ------------------------------------------------------------
plt.figure(figsize=(12,6))
plt.plot(timestamps, y_true, label='Observedo', linewidth=2)
plt.plot(timestamps, y_pred, label='Previsto', linewidth=2, alpha=0.8)
plt.title(f'Predição RNN - gmu_level_mH2O_compensado\nR² = {r2:.3f} | MAE = {mae:.3f}')
plt.xlabel('Data')
plt.ylabel('Nível (mH2O)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 7. Plotar evolução da loss
# ------------------------------------------------------------
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'], label='Training Loss', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
plt.title('RNN Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Calcular erro base
# ------------------------------------------------------------
y_pred_base = model.predict(X_test)
base_mse = mean_squared_error(y_test, y_pred_base)

# ------------------------------------------------------------
# 2. Calcular importância por permutação
# ------------------------------------------------------------
importances = []
feature_names = X.columns

for i, col in enumerate(feature_names):
    X_test_perm = X_test.copy()
    for sample in range(X_test.shape[0]):
        X_test_perm[sample, :, i] = np.random.permutation(X_test_perm[sample, :, i])
    np.random.shuffle(X_test_perm[:, i])  # embaralha a coluna i
    y_pred_perm = model.predict(X_test_perm)
    perm_mse = mean_squared_error(y_test, y_pred_perm)
    importance = perm_mse - base_mse
    importances.append(importance)

importances = np.array(importances)

# ------------------------------------------------------------
# 3. Calcular porcentagem de importância
# ------------------------------------------------------------
total_importance = np.sum(importances)
percent_importance = (importances / total_importance) * 100

# Organizar por ordem decrescente
sorted_idx = np.argsort(importances)[::-1]

# ------------------------------------------------------------
# 4. Plotar importâncias
# ------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.barh(range(len(feature_names)), percent_importance[sorted_idx][::-1])
plt.yticks(range(len(feature_names)), feature_names[sorted_idx][::-1])
plt.title("Importância das Features (Permutação)")
plt.xlabel("Importância relativa (%)")
plt.tight_layout()
plt.show()

# ------------------------------------------------------------
# 5. Mostrar top variáveis
# ------------------------------------------------------------
print("\nTop variáveis por importância (%):")
for i in sorted_idx[:10]:
    print(f"{feature_names[i]}: {percent_importance[i]:.2f}% (ΔMSE = {importances[i]:.6f})")
```
