```python
pocos_chuva_pressao = pd.read_csv('nivel-pocos_chuva_pressao_tratados.csv')
mare_terrestre = pd.read_csv('earth_tide.csv').rename(columns={'datetime': 'date_time'})

funcionamento_pocos = pd.read_csv('funcionamento_pocos_30min.csv', sep=',').dropna()
funcionamento_pocos[funcionamento_pocos.columns.difference(['date_time'])] = funcionamento_pocos[funcionamento_pocos.columns.difference(['date_time'])].astype(int)

# --- Merge dos DataFrames ---
data = pocos_chuva_pressao.merge(funcionamento_pocos, on='date_time', how='outer')
data = data.merge(mare_terrestre, on='date_time', how='outer')

# --- Ordenar pelo datetime ---
data = data.sort_values('date_time').reset_index(drop=True)
```

```python
import pandas as pd

# --- Garantir datetime ---
data['date_time'] = pd.to_datetime(data['date_time'])
data = data.set_index('date_time').sort_index()

# --- Função para encontrar períodos contínuos com dados ---
def find_continuous_periods(series, min_length):
    """
    Retorna uma lista de tuplas (start, end) para períodos contínuos com dados não NaN.
    """
    is_valid = series.notna().astype(int)
    group = (is_valid.diff(1) != 0).cumsum() * is_valid
    
    periods = []
    for g in group.unique():
        if g == 0:
            continue
        grp_idx = group[group == g].index
        if len(grp_idx) >= min_length:
            periods.append((grp_idx[0], grp_idx[-1]))
    return periods

# --- Parâmetros ---
min_records = 25 * 48  # 25 dias × 48 registros por dia

# --- Encontrar períodos para cada coluna ---
columns_to_check = [
    'gmu_level_mH2O_compensado',
    'grafica_level_mH2O_compensado',
    'correios_level_mH2O_compensado'
]

continuous_periods = {}
for col in columns_to_check:
    continuous_periods[col] = find_continuous_periods(data[col], min_records)

# --- Exibir períodos ---
for col, periods in continuous_periods.items():
    print(f"\nColuna: {col}")
    for start, end in periods:
        print(f"De {start} até {end} ({(end-start).days+1} dias)")
```

```python
import pandas as pd

# --- 0. Copia a coluna para não alterar o df original ---
level = data['gmu_level_mH2O_compensado'].copy()

# --- 1. Inicializa lista para armazenar estatísticas ---
annual_list = []

# --- 2. Itera sobre cada ano ---
for year, group in level.groupby(level.index.year):
    # Remove dados inconsistentes
    group = group[(group >= 0) & (group <= 100)]
    
    # Remove outliers usando IQR dentro do ano
    Q1 = group.quantile(0.25)
    Q3 = group.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    group_clean = group[(group >= lower_bound) & (group <= upper_bound)]
    
    # Calcula estatísticas do ano
    stats = group_clean.agg(['mean', 'min', 'max', 'std', 'median', 'count'])
    stats['amplitude'] = stats['max'] - stats['min']
    stats['cv'] = stats['std'] / stats['mean']
    
    # Adiciona o ano como coluna temporária
    stats['year'] = year
    annual_list.append(stats)

# --- 3. Cria DataFrame final ---
annual_stats = pd.DataFrame(annual_list).set_index('year')

# --- 4. Exibe resultados ---
print("Análise anual do nível do poço GMU (tratando cada ano individualmente):")
annual_stats.to
```
